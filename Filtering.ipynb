```python
import pandas as pd
import matplotlib.pyplot as plt
from highlight_text import fig_text
import matplotlib as mpl
from mplsoccer.pitch import Pitch

# Load data from CSV file
df = pd.read_csv('C:\\Users\\USER\\Documents\\Football Analytics\\IKD_REMO_B.csv')
df
```

```python
# Define a list of actions related to ground duels
#store that list in the variable called ground_duels
ground_duels = ['Ground Duel Won', 'Ground Duel Lost'] #this is a list


# Create a new DataFrame gd_df that filters the original DataFrame (df)
# to include only the rows where the 'Action' column matches any of the values in the ground_duels list
gd_df = df[df['Action'].isin(ground_duels)]
gd_df
```

```python
text_color = 'white'  


fig, ax = plt.subplots(figsize=(13,8.5))
fig.set_facecolor('grey')
ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_type='statsbomb', orientation='vertical',
              pitch_color='grass', stripe=True, line_color='#c7d5cc', figsize=(13, 8),
              constrained_layout=False, tight_layout=True)

pitch.draw(ax=ax)

#I invert the axis to make it so I am viewing it how I want
plt.gca().invert_yaxis()

# Filter data for 'Ground duel won' and 'Ground duel lost'
ground_duel_won = gd_df[gd_df['Action'] == 'Ground Duel Won']
ground_duel_lost = gd_df[gd_df['Action'] == 'Ground Duel Lost']

plt.scatter(ground_duel_won['x'], ground_duel_won['y'], s=100, c='green', alpha=.7, label='Successful Ground Duel')
plt.scatter(ground_duel_lost['x'], ground_duel_lost['y'], s=100, c='red', alpha=.7, label='Unsuccessful Ground Duel')


# Add text to the figure 
plt.title('Ikorodu City vs Remo Stars B Ground Duels\nSuccessful Ground Duel: Green | Unsuccessful Ground Duel: Red', 
          color='white', size=18)


```

```python
# Define a list of actions related to fouls committed and received
# store that list in the variable called fouls
fouls = ['Foul Committed', 'Foul Received']


# Create a new DataFrame fouls_df that filters the original DataFrame (df)
# to include only the rows where the 'Action' column matches any of the values in the fouls list
fouls_df = df[df['Action'].isin(fouls)]
fouls_df
```

```python
text_color = 'white'  


fig, ax = plt.subplots(figsize=(13,8.5))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_type='statsbomb', orientation='vertical',
              pitch_color='grass', stripe=True, line_color='#c7d5cc', figsize=(13, 8),
              constrained_layout=False, tight_layout=True)

pitch.draw(ax=ax)

#I invert the axis to make it so I am viewing it how I want
plt.gca().invert_yaxis()

# Filter data for 'Foul Committed' and 'Foul Received'
fouls_committed = fouls_df[fouls_df['Action'] == 'Foul Committed']
fouls_received = fouls_df[fouls_df['Action'] == 'Foul Received']

plt.scatter(fouls_committed['x'], fouls_committed['y'], s=100, c='green', alpha=.7, label='Foul Committed')
plt.scatter(fouls_received['x'], fouls_received['y'], s=100, c='red', alpha=.7, label='Foul Received')


# Add text to the figure 
plt.title('Foul Commited: Green | Foul Received: Red', 
          color='white', size=15)


plt.show()
```
